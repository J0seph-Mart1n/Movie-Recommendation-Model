Importing ast module, used for processing python abstract syntax grammar 

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert)
```

Fetching the director for each film

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_df['crew_director'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df
```

```python
movies_df.drop(axis=1,columns='crew',inplace=True)
```

```python
movies_df
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies_df['crew_director'] = movies_df['crew_director'].apply(lambda x:[i.replace(' ','') for i in x])
```

Vectorizing tags column (Converting text into vectors)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_movies_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_movies_df['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

PorterStemmer removes the suffixes present in the english word

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(stemming)
```

Cosine Similarity measures the similarity of two vectors.<br/>
It measures by checking the cosine of the angle between two vectors and determines if both the vectors are pointing at the same direction.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

This sorted list shows the relevancing of movie avatar compared to other movies

```python
sorted(similarity[0], reverse=True)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_movies_df[new_movies_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_movies_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```
